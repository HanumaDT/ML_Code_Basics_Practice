```python
df_require = ['Suburb', 'Rooms', 'Type', 'Price', 'Method', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Regionname', 'Propertycount'] 
```

```python
df_new = df[df_require]
df_new
```

```python
df.nunique()
```

```python
df_new.isna().sum()
```

```python
k = ['Bedroom2', 'Bathroom', 'Car']
df_new[k] = df_new[k].fillna(0)
df_new
```

```python
df_new.isna().sum()
```

```python
me = ['Landsize', 'BuildingArea', 'YearBuilt']
for i in me:
    df_new[i] = df_new[i].fillna(df_new[i].mean())
df_new
```

```python
df_new.isna().sum()
```

```python
df_new.dropna(inplace = True)
df_new.isna().sum()
```

```python
df_new.info()
```

```python
df_new = pd.get_dummies(df_new, drop_first = True).astype('int')
df_new.info()
```

```python
df_new.head()
```

```python
y = df_new['Price']
X = df_new.drop(['Price'], axis = 1)
```

```python
len(X)
```

```python
len(y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lg = LinearRegression()
```

```python
lg.fit(X_train, y_train)
```

```python
z = lg.predict(X_test)
```

```python
lg.score(X_train, y_train)
```

```python
lg.score(X_test, y_test)
```

```python
from sklearn.linear_model import Lasso
llm = Lasso(alpha=50, max_iter=100, tol=0.1)
llm.fit(X_train, y_train)
```

```python
llm.predict(X_test)
```

```python
llm.score(X_test, y_test)
```

```python
llm.score(X_train, y_train)
```

```python
from sklearn.linear_model import Ridge
rlm = Ridge(alpha=50, max_iter=100, tol=0.1)
rlm.fit(X_train, y_train)
```

```python
rlm.predict(X_test)
```

```python
rlm.score(X_train, y_train)
```
