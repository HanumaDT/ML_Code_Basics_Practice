```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("Downloads/carprices.csv")
df
```

```python
plt.scatter(df['Mileage'], df['Sell Price($)'], color='red', marker= '*')
plt.xlabel('Mileage')
plt.ylabel('Sell Price')
plt.show()
```

```python
plt.scatter(df['Age(yrs)'], df['Sell Price($)'])
plt.show()
```

```python
y = df['Sell Price($)']
X = df.drop(['Sell Price($)'], axis = 1)
from sklearn.model_selection import train_test_split
X_train, X_text, Y_train, Y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, Y_train)
```

```python
reg.predict(X_text)
```

```python
Y_test
```

```python
reg.score(X_train, Y_train)
```

```python
reg.score(X_text, Y_test)
```
