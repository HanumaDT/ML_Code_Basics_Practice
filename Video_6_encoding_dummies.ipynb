```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Downloads/Home_Price.csv")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis = 1)
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis = 1)
final
```

```python
final['monroe township'] = final['monroe township'].map({True:1, False:0})
final['robinsville'] = final['robinsville'].map({True:1, False:0})
final
```

```python
model = linear_model.LinearRegression()
model.fit (final[['area', 'monroe township', 'robinsville']], final.price)
```

```python
model.predict([[2800, 0, 1]])
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model.predict([[3000, 0, 0]])
```

```python
x = final[['area', 'monroe township', 'robinsville']]
y = final['price']
model.score(x,y)
```
