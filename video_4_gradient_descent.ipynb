```python
import numpy as np
```

```python
def gradient_descent(x, y):
    slope = intercept = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.08
    for i in range(iterations):
        y_cal = slope*x + intercept
        cost = 1/n*sum(val**2 for val in (y - y_cal))
        slope_d = -(2/n)*sum(x*(y - y_cal))
        intercept_d = -(2/n)*sum(y - y_cal)
        slope = slope - learning_rate*slope_d
        intercept = intercept - learning_rate*intercept_d
        print("iteration:", i, slope, intercept, cost)
        

a = np.array([1, 2, 3, 4, 5])
b = np.array([5, 7, 9, 11, 13])
gradient_descent(a, b)
        
    
```
