```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(iris['data'], columns=iris.feature_names)
df
```

```python
df['flower'] = iris['target']
```

```python
df
```

```python
df['flower'] = df['flower'].apply(lambda x: iris['target_names'][x])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], test_size = 0.25)
```

```python
from sklearn.svm import SVC
svm = SVC(C=10, kernel = 'rbf', gamma = 'auto')
```

```python
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(C = 10, kernel = 'rbf', gamma = 'auto'), iris['data'], iris['target'], cv = 5)
```

```python
kernels = ['linear', 'rbf', 'poly']
C_val = [1, 5, 10, 20, 30, 40, 50]
avg_scores = {}
for i in kernels:
    for j in C_val:
        print (i, j)
        cv_results = cross_val_score(SVC(C=j, kernel = i, gamma = 'auto'), iris['data'], iris['target'], cv = 5)
        avg_scores[i+'_'+str(j)] = np.average(cv_results)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
gsc = GridSearchCV(SVC(gamma = 'auto'), {
    'C':[1, 5, 10, 20, 30, 40, 50],
    'kernel':['rbf', 'poly', 'linear']
}, cv = 5, return_train_score=False)
gsc.fit(iris['data'], iris['target'])
gsc.cv_results_
```

```python
df_cv = pd.DataFrame(gsc.cv_results_)
df_cv
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
parameters = {'logistic_regression':
              {'model': LogisticRegression(solver = 'liblinear', multi_class='auto'),
                'params': {'C':[1, 5, 10]}},
              'svm' : 
                {'model': SVC(gamma = 'auto'),
                'params' : {'C':[1, 5, 10, 20, 50],
                            'kernel':['rbf', 'linear', 'poly']}},
              'random_forest': 
                  {'model':RandomForestClassifier(), 
                'params': {'n_estimators': [1, 5, 10, 25, 50]}}}
```

```python
scores = []
for model_name, mp in parameters.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris['data'], iris['target'])
    scores.append({'model': model_name, 'best_score': clf.best_score_, 'best_parameters': clf.best_params_})
scores
```
