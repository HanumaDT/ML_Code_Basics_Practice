```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits
```

```python
df = pd.DataFrame(digits['data'])
df
```

```python
df['target'] = digits['target']
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3)
```

```python
from sklearn.model_selection import train_test_split
def model_scores(model, X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
y = df['target']
X = df.drop(['target'], axis = 1)
model_scores(LogisticRegression(), X, y)
```

```python
model_scores(SVC(), X, y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from statistics import mean
mean(cross_val_score(LogisticRegression(), X, y))
```

```python
mean(cross_val_score(SVC(), X, y))
```

```python
mean(cross_val_score(RandomForestClassifier(n_estimators=40), X, y))
```

```python
scores_lg = []
scores_svc = []
scores_rfc = []
for train_index, test_index in kf.split(digits.data):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], digits.target[train_index], digits.target[test_index]
    LogisticRegression().fit(X_train, y_train)
    SVC().fit(X_train, y_train)
    RandomForestClassifier().fit(X_train, y_train)
    scores_lg.append(LogisticRegression.score(X_test, y_test))
    scores_svc.append(SVC.score(X_test, y_test))
    scores_rfc.append(RandomForestClassifier.score(X_test, y_test))
```
